# Contexto

Los números irracionales son una base importante de las matemáticas, pues permiten hacer cosas como calcular el largo de la diagonal de un cuadrado de lado 1, lo que da origen a las raíces cuadradas con las cuales estamos familiarizados por el teorema de Pitágoras.
Para poder hacer cálculos que involucren los números irracionales antiguamente se ocupaban tablas con aproximaciones de los números. Una forma de obtener estas tablas es mediante iteraciones de punto fijo, por ejemplo, para el calculo de $\sqrt{2}$ en el libro guía (ejemplo 1.6) se presenta la siguiente IPF:

$$x_{i+1} = \frac{x_i+\frac{2}{x_i}}{2}$$

Esta IPF data del siglo I d.C. y fue usada para calcular el valor de $\sqrt{720}$.

En nuestro caso, nos interesa lograr algo similar pero para el logaritmo natural, el cual cumple varias propiedades interesantes, además de guardar una fuerte relación con la constante $e$, una de las constantes más importantes del mundo de las matemáticas.

## Funciones Entregadas
*No debe modificarlas*

```python
def IPF(x0, g, tol):
    """
    Parameters
    ----------
    x0        : float
                Initial value
    g         : function
                Function to use in the IPF
    tol       : float
                Max error

    Returns
    -------
    x      : float
             value after convergence
    """
    x = x0

    while abs(x - g(x)) > tol:
        x = g(x)

    return x

def Bisection(a, b, f, tol):
    """
    Parameters
    ----------
    a         : float
                Interval start
    b         : float
                Interval end
    f         : Function
                Funtion to evaluate
    tol       : float
                Max error

    Returns
    -------
    val   : float
            Root for the function f with tolerance tol
    """
    while((b-a)/2  > tol):
        c = (a+b)/2

        if f(c) == 0:
            break
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    val = c
    return val
```

Como:

$log(2) = x$

Se tiene que buscar un punto fijo tal que g'(x) < 1,
como no se puede utilizar np.log() se reescribe la funcion como:

$e^x-2=0$

Ahora se trabaja la funcion para encontrar g(x);

$e^x-2=0$ /+1

$e^x-1=1$ /*x

$x*e^x-x=x$

$x = \frac{x}{e^x-1}$

Con esto se tiene:
$g(x) = \frac{x}{e^x-1}$

**[2 puntos] Pregunta 1.2:** Impleméntela en código y obtenga el valor con al menos 5 decimales de precisión.

```python
# Iteracion de punto fijo
def FPI_log2(x):
    """
    Parameters
    ----------
    x         : float
                Initial value

    Returns
    -------
    val   : float
             Numerical aproximation of ln(2)
    """
    f = lambda x: x/(np.exp(x)-1)
    val = IPF(x, f, 1e-6)
    return val

FPI_log2(2)
```

De la misma forma que la pregunta anterior log(a) = x se puede reescribir como:

```python
f = lambda x,num: np.exp(x) - num

def log_aprox(a,b,f,num):
    """
    Parameters
    ----------
    a         : float
                Interval start
    b         : float
                Interval end
    f         : Function
                Funtion to evaluate
    num       : float
                Value to aproximate, log(num)
    Returns
    -------
    val   : float
             Numerical aproximation of log(num)
    """
    #biseccion de 2 variables a y x
    def biseccion2var(a, b, f, num, tol):
        while((b-a)/2  > tol):
            c = (a+b)/2
            if f(c, num) == 0:
                break
            if f(a, num)*f(c, num) < 0:
                b = c
            else:
                a = c

        val = c
        return val

    val = biseccion2var(a, b, f, num, 0.5e-6)
    return val
    

# Valor de prueba
log_aprox(2,3,f,15)
```

```python
def log_vec(x, tol):
    """
    Parameters
    ----------
    x              : ndarray
                    Values to aproximate
    tol            : float
                    Max error
    Returns
    -------
    val   : float
             Numerical aproximation of ln(num)
    """
    #no dse especifica ocupar ningun metodo en especifico asi que ocupo el de biseccion
    def biseccion2var(a, b, f, num, tol):
        while((b-a)/2  > tol):
            c = (a+b)/2
            if f(c, num) == 0:
                break
            if f(a, num)*f(c, num) < 0:
                b = c
            else:
                a = c

        val = c
        return val

    x2 = []
    for n in x:
        x2.append(biseccion2var(0, 2, f, n, 1e-6))
    x = np.array(x2)
    return x

# Valor de prueba
x0 = np.array([3,4,5])
log_vec(x0, 1e-6)
```

**[10 puntos] Pregunta 5.1:** Coloque su desarrollo para obtener la función $g(x)$ de la IPF a iterar haciendo uso de latex en markdown.

$3^x\times4^{x+3}-5=0$   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;/+5

$3^x\times4^{x+3}=5$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /log()

$log(3^x\times4^{x+3})=log(5)$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /Aplicando propiedades de logaritmos

$log(3^x) + log(4^{x+3})=log(5)$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /y mas propiedades de logaritmos

$x*log(3) + x*log(4) + 3*log(4)=log(5)$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /reordenando

$x*(log(3)+log(4)) =log(5) - 3*log(4)$

Haciendo el siguiente cambio de variable para facilitar el desarrollo:

$A = log(3) + log(4)$

$B = log(5) - 3*log(4)$

Ahora se busca el punto fijo de la funcion:

$Ax+B=0$

$Ax+B=0$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /+$x$

$Ax+B+x=x$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; /-$B$

$x(A+1)=x - B$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

$x = \frac{x+B}{A+1}$

Con esto se termina:
$g(x)= \frac{x-B}{A+1}$

**[10 puntos] Pregunta 5.2:** Impleméntela en código

```python
def solve(x0, tol=1e-6):
    """
    Parameters
    ----------
    x0        : float
                Initial Value
    tol       : float
                Max error
    Returns
    -------
    val    : float
            Root that solves the equation
    """
    #calculate logs
    log3, log4, log5 = log_vec(np.array([3,4,5]), 1e-6)
    A = log3 + log4
    B = (3*log4) - log5
    f = lambda x: (x-B)/(A+1)
    val = IPF(x0, f, tol)
    return val
    
solve(-1)
```
